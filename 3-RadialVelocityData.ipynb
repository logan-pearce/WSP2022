## Computing planet mass from RV signals:
The physics of computing the mass via RV is the same as the seesaw actually.  The equation is given by Newton's law of gravitation:

$$F_g = G \frac{M_* M_p}{r^2}$$

Where $M_*$ and $M_p$ are mass of star and planet, $G$ is the gravitational constant (same as above), and $r$ is the semi-major axis determined from the period, and $F_g$ is the force of gravity between star and planet.

To get the force of gravity, we need to use Newton's second law of motion:
$$ F_g = ma$$
where $m$ is mass (in this case it would be the total mass $m$ = $M_*$ + $M_p$) and $a$ is acceleration.  Acceleration is defined as change in velocity over time:
$$ a = \frac{v_2 - v_1}{t_2 - t_1}$$

There is no way to know the inclination of the orbit from the RV signal alone.  The RV signal is the velocity projected along our line of sight only, but the planet has velocity in three dimensions.  So the mass we determine from an RV signal alone is only a <font color='cornflowerblue'> **Minimum Mass**</font>!

True mass: $$m_p = \frac{m_{rv}}{sin(i)} $$

So RV masses are usually reported as $M_p$ sin(i)
